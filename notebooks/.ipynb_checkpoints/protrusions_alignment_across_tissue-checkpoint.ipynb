# Measure tissue-wide protrusion polarity (several versions)

* Uses protrusion base-to-tip angles measured from timelapse movies of membrane-dyed epithelia (see `../code/measure_hemijunction_traits.py`) and the net migration direction calculated for the timelapse (`..../code/measure_migration_speed.py`).

* For protrusion-protrusion alignment, calculates the dot product of the orientations of all pairs of protrusions present at the same timepoint. Then takes their mean as a tissue-wide, timelapse-wide metric of protrusion alignment. 

* For protrusion-migration alignment, calculates the dot product of each protrusion orientation and the direction of tissue migration, then takes their mean. 

* To distinguish between (1) high axial polarity but low vectorial polarity vs (2) low axial polarity polarity, versions of the tissue metric are also calculated using the absolute values of the individual prot-prot or prot-mig dot products. In the case of high axial, low vectorial polarity, the non-abs metrics will be low but these will be high. 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
from pathlib import Path
import pandas as pd

sys.path.append("../code/")
from functions.utils import (select_files,
                             cart_to_pol,
                             pol_to_cart,
                             make_and_store_prot_dfs,
                             store_hjs_in_dataset_dicts)
```

## Set data, output locations

```python
hjs_dir = "../data/membrane_protrusivity_polarity/sample_data/"
mig_dir = "../data/membrane_protrusivity_polarity/"
out_dir = "../data/membrane_protrusivity_polarity/"

mig_path = os.path.join(mig_dir, "migration_speed_sample.csv")
out_path = os.path.join(out_dir, "prot_alignment_sample.csv")
```

## Set plot params

```python
palette = dict(w1118 = "#2631ED",
               fat2N1032 = "#1AC1BC",
               lar132bola1 = "#EA0A0A",
               sema5cK175 = "#FFA500",
               TJ_larRNAi_sema5cK175 = "#CC33CC",
               ck666 = "#7B7B7B") # Not plotted in paper
```

## Import migration speeds and hjs, filter out non-protrusive hjs

```python
# Import and filter hemijunctions data, store
datasets = select_files(hjs_dir, "_data_hjs.csv")
datasets = store_hjs_in_dataset_dicts(datasets)

for dataset in datasets:
    df_hjs = dataset["df_hjs"]
    dataset["df_hjs"] = df_hjs

datasets = make_and_store_prot_dfs(datasets)

# Import and process migration data, store
df_mig = pd.read_csv(mig_path, index_col=0)

for dataset in datasets:
    basename = dataset["basename"]
    df_hjs = dataset["df_hjs"]
    df_mig_sample = df_mig[df_mig["sample"] == basename]
    dataset["df_mig_sample"] = df_mig_sample
```

## Calculate dot products of migration and protrusion vectors
* Dot product of 2 unit vectors will be between -1 (opposite direction) and 1 (same direction) (0 = orthogonal).
* Find the average of all dot products (the metric for alignment with migration)
* Find the average of the absolute value of all dot products (metric for aligment with the migration axis, either in the same or opposite direction)

```python
alignment_dicts = []

for dataset in datasets:
    condition = dataset["condition"]
    sample_num = dataset["sample_num"]
    basename = dataset["basename"]
    
    df_mig_sample = dataset["df_mig_sample"]
    df_prots = dataset["df_prots"]
    mig_speed = df_mig_sample.iloc[0]["mig_speed_um_per_min"]

    # Get migration direction (cartesian)
    mig_dir_rad = df_mig_sample["mig_dir_rad"].iloc[0]
    mig_x, mig_y = pol_to_cart(mig_dir_rad, 1)
    
    # Get protrusion directions (cartesian)
    prot_arr = np.zeros((len(df_prots), 2))
    for i in range(len(df_prots)): 
        prot_dir_rad = df_prots.iloc[i]["prot_angle_rad"]
        prot_arr[i,:] = pol_to_cart(prot_dir_rad, 1)
        
    # Calc protrusion * migration direction for each protrusion
    dot_arr_mig = np.zeros(len(df_prots))
    for i in range(len(prot_arr)):
        dot_arr_mig[i] = np.vdot([mig_x, mig_y], prot_arr[i])
        
    # Calc protrusion * protrusion direction (per frame)
    dot_ls = []
    for t in range(len(df_prots)):
        df_prots_frame = df_prots[df_prots["t_step"] == t]
        prot_arr = np.zeros((len(df_prots_frame), 2))
        for i in range(len(df_prots_frame)):
            prot_dir_rad = df_prots_frame.iloc[i]["prot_angle_rad"]
            prot_arr[i,:] = pol_to_cart(prot_dir_rad, 1)
        for i in range(len(prot_arr)):
            for j in range(len(prot_arr)):
                if i < j: 
                    dot = np.vdot(prot_arr[i], prot_arr[j])
                    dot_ls.append(dot)
            
    # Append info to dataframe
    alignment_dicts.append({"condition" : condition,
                            "sample_num" : sample_num,
                            "basename" : basename,
                            "mig_speed_um_per_min" : mig_speed,
                            "prot_x_prot_dot" : np.mean(dot_ls),
                            "prot_x_prot_dot_abs" : np.mean(np.abs(dot_ls)),
                            "prot_x_mig_dot" : np.mean(dot_arr_mig),
                            "prot_x_mig_dot_abs" : np.mean(np.abs(dot_arr_mig)),
                           })
    
df_alignment = pd.DataFrame(alignment_dicts)
```

## Plot results
Note: Only one sample from each condition is plotted here. The full dataset of alignment measurements can be found in the `membrane_protrusivity_polarity` folder.

```python
# Plot protrusion x migration alignment
x = "mig_speed_um_per_min"
y = "prot_x_prot_dot"

ax = sns.scatterplot(data=df_alignment, x=x, y=y, hue="condition", palette=palette)
ax.set_xlabel("Migration speed (um per min)")
ax.set_ylabel(y)
ax.set_title("Prot-prot alignment vs migration speed")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
# Plot protrusion x migration alignment
x = "mig_speed_um_per_min"
y = "prot_x_mig_dot"

ax = sns.scatterplot(data=df_alignment, x=x, y=y, hue="condition", palette=palette)
ax.set_xlabel("Migration speed (um per min)")
ax.set_ylabel(y)
ax.set_title("Prot-mig alignment vs migration speed")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
# Plot protrusion x migration alignment
x = "prot_x_prot_dot"
y = "prot_x_mig_dot"

ax = sns.scatterplot(data=df_alignment, x=x, y=y, hue="condition", palette=palette)
ax.set_xlabel("Migration speed (um per min)")
ax.set_ylabel(y)
ax.set_title("Prot-prot alignment vs prot-mig alignment")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
# Plot protrusion x migration alignment
x = "prot_x_prot_dot"
y = "prot_x_prot_dot_abs"

ax = sns.scatterplot(data=df_alignment, x=x, y=y, hue="condition", palette=palette)
ax.set_xlabel("Migration speed (um per min)")
ax.set_ylabel(y)
ax.set_title("Prot-prot vectorial alignment vs prot-prot axial alignment")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

## Output tissue protrusion aligment df as csv

```python
df_alignment = df_alignment.sort_values("basename")
df_alignment.to_csv(path_or_buf = out_path)
```
