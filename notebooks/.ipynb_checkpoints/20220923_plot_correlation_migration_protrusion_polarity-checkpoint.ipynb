Goal: lar13.2 and sema5cK175 epithelia both migrate slowly, but only lar has these dramatic off-migration-axis protrusions. But sema5cK175 is complicated- migration recovers over time, and migration speed seems to correlate decently with protrusion polarity. Plot protrusion polarity as a function of migration speed and see if lar and sema5c are distinguishable. 

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
DATA_DIR = "/Volumes/audrey2/Data_Analysis/CellMask_genoPanel_signalosome/"

mig_path = os.path.join(DATA_DIR, "migration_speed.csv")
prot_pol_path = os.path.join(DATA_DIR, "mean_prot_avg_len.csv")

df_mig = pd.read_csv(mig_path, index_col=0)
df_prot = pd.read_csv(prot_path, index_col=0)
```

## Combine mig and prot data for samples with both

```python
mig_samples = df_mig["sample"].unique()
prot_samples = df_prot["sample"].unique()
shared_samples = list(set(mig_samples) & set(prot_samples))

df = pd.DataFrame(columns = ['condition', 'sample', 'mig_speed', 'prot_len'])

for sample in shared_samples:
    condition = df_prot[df_prot['sample']==sample].iloc[0]['condition']
    mig_speed = df_mig[df_mig['sample']==sample].iloc[0]['mig_speed_um_per_min']
    prot_len = df_prot[df_prot['sample']==sample].iloc[0]['prot_avg_len_mean_um']
    
    df = df.append({'condition':condition,
                    'sample':sample,
                    'mig_speed':mig_speed,
                    'prot_len':prot_len}, 
                ignore_index = True)
    
df.sort_values(by='sample', axis=0, inplace=True)
df
```

```python
colors = {'w1118':'tab:blue',
          'fat2N1032':'tab:orange',
          'ck666':'tab:green',
          'fat2dICD':'tab:red',
          'lar132bola1':'tab:purple',
          'sema5cK175':'tab:brown',
          'TJ_larRNAi_sema5cK175':'tab:pink'
         }

fig, ax = plt.subplots()
ax.scatter(df["prot_len"], df["mig_speed"], c=df['condition'].map(colors))
ax.set_ylabel("migration speed (um/min)", fontsize=14)
ax.set_xlabel("avg protrusion len (um)", fontsize=14)

# add a legend
from matplotlib.lines import Line2D
handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=v, label=k, markersize=8) for k, v in colors.items()]
ax.legend(title='color', handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')
```
